```python
from smolagents import CodeAgent, DuckDuckGoSearchTool, tool, LiteLLMModel
import wikipediaapi

# Naya model
model = LiteLLMModel(
    model_id="groq/llama-3.1-8b-instant",
    api_key="YOUR_GROQ_API_KEY_HERE"
)

@tool
def wikipedia_search(query: str) -> str:
    """
    Searches Wikipedia for information.
    Args:
        query: Topic to search
    """
    wiki = wikipediaapi.Wikipedia(
        user_agent='GAIA-AI-Agent/1.0',
        language='en'
    )
    page = wiki.page(query)
    if page.exists():
        return page.summary[:1000]
    return "No Wikipedia page found"

@tool
def calculate(expression: str) -> str:
    """
    Calculates math expression.
    Args:
        expression: Math like '2+2'
    """
    try:
        return str(eval(expression))
    except:
        return "Invalid expression"

agent = CodeAgent(
    tools=[DuckDuckGoSearchTool(), wikipedia_search, calculate],
    model=model,
    max_steps=5
)

questions = [
    "Who invented the telephone?",
    "What is 144 multiplied by 5?"

]

for i, question in enumerate(questions):
    print(f"\n--- Question {i+1} ---")
    print(f"Q: {question}")
    try:
        answer = agent.run(question)
        if answer:
            print(f"A: {answer}")
        else:
            print("A: (No answer returned)")
    except Exception as e:
        print(f"Error: {e}")

print("\nAll done!")
```
